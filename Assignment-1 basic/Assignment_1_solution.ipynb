## Exercise 2
Using scatter plot:  
a) Show the relation of square footage and price.  
b) Check how price is influenced by apartment's grade. Find other variables correlated with price.  
c) Check how good apartments are distributed over the city. First, plot zipcode versus price. Then, use apartments coordinates (longitude and lattitude) in order to show where apartments of good grade are located. 

```python
#a)
#Your code:

plt.scatter(data.sqft_living, data.price)
plt.xlabel("Square footage")
plt.ylabel("Price")

#Hint plt.scatter
```

> As we can see square footage and price are correlated weakly. Bigger living spaces costs more but apartment's location is also influences the price.

```python
#b)
#Your code:

plt.scatter(data.grade, data.price)
plt.xlabel("Grade")
plt.ylabel("Price")
```

> As the grade increases, the price increases aswell

```python
plt.scatter(data.sqft_lot, data.price)
plt.xlabel("Square footage of land space")
plt.ylabel("Price")
```

> Square footage of land space to price is correlated weakly, negative and non linear.

```python
plt.scatter(data.bathrooms, data.price)
plt.xlabel("Number of bathrooms")
plt.ylabel("Price")
```

> Square footage of living space and number of bathrooms are also correlated to price but weakly.

```python
#c)
#Your code:

plt.scatter(data.zipcode, data.price)
plt.xlabel("Zipcode")
plt.ylabel("Price")
```

> Using this chart, we can see the outline of the division into districts and indicative house prices in them

```python
#Your code:

plt.figure(figsize = (13, 11), dpi = 80)
plt.scatter(data.long, data.lat, c = data.price, cmap = "Greens_r")
plt.xlabel("Longtitude")
plt.ylabel("Latitude")
```

Most expensive apartments are located in city center.

## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:

# df2=df.loc[(df['Discount'] >= 1000) & (df['Discount'] <= 2000)]

data_bbf = data.loc [(data['bedrooms'] == 3) & (data['bathrooms'] == 2) & (data['floors'] == 2)]
data_bbf.describe()

#Hint: pandas.DataFrame.loc
```

```python
#b)
#Your code:

plt.scatter(data.grade, data.price)
plt.scatter(data_bbf.grade, data_bbf.price)
plt.xlabel("Grade")
plt.ylabel("Price")
```

```python
#c)
#Your code:

data_bbf['price_to_footage'] = data_bbf['price'] / data_bbf['sqft_living']

data_bbf.sort_values('price_to_footage', ascending=False)

#Hint: .loc, .sort_values(), .index
```
