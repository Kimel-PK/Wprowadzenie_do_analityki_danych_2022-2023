> Note: As the grade increases, the price increases aswell.
It is worth notice that `grade` are discrete values, so we can observe characteristic vertical lines on X axis.

```python
plt.scatter(data.sqft_lot, data.price)
plt.xlabel("Square footage of land space")
plt.ylabel("Price")
```

```python
plt.scatter(data.bathrooms, data.price)
plt.xlabel("Number of bathrooms")
plt.ylabel("Price")
```

```python
#c)
#Your code:

plt.scatter(data.zipcode, data.price)
plt.xlabel("Zipcode")
plt.ylabel("Price")
```

```python
#Your code:

plt.figure(figsize = (13, 11), dpi = 80)
plt.scatter(data.long, data.lat, c = data.price, cmap = "Greens_r")
plt.xlabel("Longtitude")
plt.ylabel("Latitude")
```

> Note: Most expensive apartments are located in city center.

## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:

# df2=df.loc[(df['Discount'] >= 1000) & (df['Discount'] <= 2000)]

data_bbf = data.loc [(data['bedrooms'] == 3) & (data['bathrooms'] == 2) & (data['floors'] == 2)]
data_bbf.describe()

#Hint: pandas.DataFrame.loc
```

```python
#b)
#Your code:

plt.scatter(data.grade, data.price)
plt.scatter(data_bbf.grade, data_bbf.price)
plt.xlabel("Grade")
plt.ylabel("Price")
```

> Note: Searched apartments are located betveen sixth add tenth grade inclusively

```python
#c)
#Your code:

data_bbf = data_bbf.copy()
data_bbf['price_to_footage'] = data_bbf['price'] / data_bbf['sqft_living']

data_bbf.sort_values('price_to_footage', ascending=False)

#Hint: .loc, .sort_values(), .index
```

> Note: The most fitting apartment with the best price to footage ratio is apartment with ID 9608
